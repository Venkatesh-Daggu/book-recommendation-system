```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books=pd.read_csv('Books.csv',low_memory=False)
books.head()
```

```python
books.info()
```

```python
books.columns
```

```python
books=books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher',
       'Image-URL-L']]
```

```python
books.head(2)
```

```python
books.rename(columns={
    'Book-Title': 'Title',
    'Book-Author':'Author',
    'Year-Of-Publication':'Year',
    'Image-URL-L':'img_url'
},inplace=True)
```

```python
books.head(2)
```

```python
users=pd.read_csv('Users.csv')
users
```

```python
ratings=pd.read_csv('Ratings.csv')
ratings
```

```python
ratings.rename(columns={
    'Book-Rating':'rating'
},inplace=True)
```

```python
x=ratings['User-ID'].value_counts()>200
x[x].shape
```

```python
y=x[x].index
y
```

```python
ratings=ratings[ratings['User-ID'].isin(y)]
ratings.shape
```

```python
rating_with_books=ratings.merge(books,on='ISBN')
rating_with_books.head(2)
```

```python
rating_with_books.shape
```

```python
num_rating= rating_with_books.groupby('Title')['rating'].count().reset_index()
```

```python
num_rating.rename(columns={'rating':'num_of_ratings'},inplace=True)
```

```python
num_rating.head()
```

```python
final_rating=rating_with_books.merge(num_rating,on='Title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['num_of_ratings']>=50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['User-ID','Title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='User-ID',index='Title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion= model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
book_pivot.index
```

```python
book_name=book_pivot.index
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
pickle.dump(book_name,open('book_name.pkl','wb'))
pickle.dump(final_rating,open('final_rating.pkl','wb'))
pickle.dump(book_pivot,open('book_pivot.pkl','wb'))
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance, suggestion= model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(suggestion)):
        books=book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name='84 Charing Cross Road'
recommend_book(book_name)
```
